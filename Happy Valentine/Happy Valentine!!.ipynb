```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# This first two list will store the points to plot a hearth
x = []
y1 = []

l = np.linspace(-2,2,500)  # 500 samples from -2 to 2

# This two list will store the points to plot a bird
x2 = []
y2 = []
```

## There are many ways to plot a hearth. In this notebook I'll use the following functions:

$f(x) = \sqrt{1-(|x|-1)^2}$


$ g(x) = arccos(1-|x|) - \pi$


```python
for i in l:
    try:
        y1.append(np.sqrt(1-(abs(i)-1)**2))
        x.append(i)
        y1.append(math.acos(1-abs(i))-np.pi)
        x.append(i)
        
    except:
        x = x[:-1]
        y1 = y1[:-1]

    x2.append((i-0.5,np.sqrt(1-(abs(i)-1)**2)))
    y2.append((i-0.5,math.acos(1-abs(i))-np.pi))
```

### Let's check the size of the previous variables

```python
print(f"len(x): {len(x)}, len(y1): {len(y1)}, len(x2): {len(x2)}, len(y2): {len(y2)}")
```

It's fine beacause the first two lists have the same shapes and this also happens with the last two lists.

```python
print(f"x:{x[:3]}\n\ny:{y1[:3]}\n\nx2:{x2[:3]}\n\ny2:{y2[:3]}")
```

```python
%matplotlib notebook

fig,ax = plt.subplots(figsize=(4,4),dpi=100)  # 400*400 picture

line, = ax.plot(x,y1,color="#ff0000",alpha=1,zorder = 0)   # hearth plot

line2, = ax.plot(x,y1,color="#cccaaa",zorder = 5)  # bird plot

#ax.set_axis_off()   # hidden the axis for beautiful purpose only

for location in ["left","right","bottom","top"]:
    ax.spines[location].set_color("w") 
#ax.set_facecolor("#022065") 
#Adding some elements to the main plot such as three lines and the eye of the bird

positions = {"first_line": [(-0.5,-2), (-2,-3)],      
            "second_line": [(-1,-0.2), (-3,-2.5)],
            "third_line": [(-1.5,-0.5), (-3,-2.3)],
            "eye": (1.04,0.9)}

# Doing a simple plot animation
def animate(num,x,y,x2,y2,line,line2):
    line.set_data(x[:num],y[:num])
    line2.set_data(x2[:num],y2[:num])
    return line,line2

ani = FuncAnimation(fig,animate,
                    len(x),
                    fargs = [x,y1,x2,y2,line,line2],
                    interval=1,
                    blit=True,
                    repeat=True)

for i in list(positions.keys())[:-1]:
    lines = plt.plot(positions[i][0], positions[i][1],"-k")
ax.tick_params(top=False,left=False,bottom=False)
ax.set_yticks([])
ax.set_xticks([])
fig.patch.set_facecolor('w')
ax.scatter(positions["eye"][0], positions["eye"][1], c="k",zorder = 10);
```

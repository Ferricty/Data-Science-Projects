```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
# Importing the libraries
# Read data from file
df=pd.read_csv("epa-sea-level.csv")

#Setting the options to None to display better the information
pd.options.display.max_rows = None
pd.options.display.max_columns = None
df
```

```python
# Create scatter plot
fig,ax=plt.subplots()
ax.scatter(x="Year",y="CSIRO Adjusted Sea Level",data=df)

# Create first line of best fit
# lineregress return five values so they are declared even not all are used
#    slope : float
#        Slope of the regression line.
#    intercept : float
#        Intercept of the regression line.
#    rvalue : float
#        The Pearson correlation coefficient. The square of rvalue is equal to the coefficient of determination.
#    pvalue : float
#        The p-value for a hypothesis test whose null hypothesis is that the slope is zero, using Wald Test with t-distribution of the test statistic. See alternative above for alternative hypotheses.
#    stderr : float
#        Standard error of the estimated slope (gradient), under the assumption of residual normality.
        
slope, intercept, r, p, se = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
y=pd.Series(range(1880,2050))

# Plot takes x and y values; 'r' indicates the color of the line
ax.plot(y,intercept+slope*y,'r',label='first line of best fit')
plt.show()
```

### Actually we have the first line of best fit. Now let's go to create the second. 

```python
# Create scatter plot
fig,ax=plt.subplots()
ax.scatter(x="Year",y="CSIRO Adjusted Sea Level",data=df)

# Create second line of best fit
df_second_line=df.loc[df["Year"]>=2000]
slope2, intercept2, r2, p2, se2 = linregress(df_second_line["Year"],df_second_line["CSIRO Adjusted Sea Level"])
y2=pd.Series(range(2000,2050))
ax.plot(y2,intercept2+slope2*y2,'k',label='second line of best fit')
plt.show()
```

### Now it's time to join both scatter plots in one picture. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress


# Read data from file
df=pd.read_csv("epa-sea-level.csv")

# Create scatter plot
fig,ax=plt.subplots()
ax.scatter(x="Year",y="CSIRO Adjusted Sea Level",data=df)

# Create first line of best fit
slope, intercept, r, p, se = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
y=pd.Series(range(1880,2050))
ax.plot(y,intercept+slope*y,'r',label='first line of best fit')
# Create second line of best fit
df_second_line=df.loc[df["Year"]>=2000]
slope2, intercept2, r2, p2, se2 = linregress(df_second_line["Year"],df_second_line["CSIRO Adjusted Sea Level"])
y2=pd.Series(range(2000,2050))
ax.plot(y2,intercept2+slope2*y2,'k',label='second line of best fit')
ax.set(title='Rise in Sea Level',xlabel="Year",ylabel="Sea Level (inches)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress


# Read data from file
df=pd.read_csv("epa-sea-level.csv")

# Create scatter plot
fig,ax=plt.subplots(1,1,figsize=(9,6.9),dpi=100)
ax.scatter(x="Year",y="CSIRO Adjusted Sea Level",data=df,alpha=0.5)

# Create first line of best fit
slope, intercept, r, p, se = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
y=pd.Series(range(1880,2050))
ax.plot(y,intercept+slope*y,'r',label='First Line of Best Fit')
# Create second line of best fit
df_second_line=df.loc[df["Year"]>=2000]
slope2, intercept2, r2, p2, se2 = linregress(df_second_line["Year"],df_second_line["CSIRO Adjusted Sea Level"])
y2=pd.Series(range(2000,2050))
ax.plot(y2,intercept2+slope2*y2,'--k',label='Second Line of Best Fit')


# Add labels and title

plt.title('Rise in Sea Level',fontsize=22,loc="left",x=0.05,y=0.93,transform=fig.transFigure,fontweight="bold",fontfamily="serif")
plt.suptitle("Changes in the sea level since 1880 and a prediction to 2050",fontsize=18,ha="left",x=0.05,y=0.9,fontfamily="serif")
ax.set_facecolor('#CCCCCC')

#xaxis properties
ax.set_xlabel("Year",ha="right",fontsize=16,fontweight="regular",fontfamily="sans")
ax.xaxis.set_label_coords(x=1,y=-0.07,transform=ax.transAxes)

#yaxis properties
ax.set_ylabel("Sea Level (inches)",ha="right",fontsize=16,fontweight="regular",fontfamily="sans")
ax.yaxis.set_label_coords(x=-0.07,y=1,transform=ax.transAxes)

ax.tick_params(axis="both",labelsize=12)
plt.subplots_adjust(top=0.82)

plt.legend()

# Save plot and return data for testing (DO NOT MODIFY)
plt.savefig('sea_level_plot.png')
```
